```python
# This Imports libraries
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This Loads dataset in CSV 
df = pd.read_csv('Dataset.csv')
# This displays general information about the dataset
print("Dataset Information:")
print(df.info())
```

```python
# This displays Summary Statistics of the numerical columns in the dataset
print("\nSummary Statistics:")
print(df.describe())
```

```python
# This display the first few rows of the dataset
print("\nFirst Few Rows:")
print(df.head())
```

```python
# This displays if dataset have any missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# This visualizes the distribution of Precipitation through Histogram
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Precipitation', color='mediumseagreen', kde=True)
plt.xlabel('Precipitation')
plt.ylabel('Frequency')
plt.title('Distribution of Precipitation')
plt.show()
```

```python
# This visualizes the distribution of Maximum Temperature through Histogram
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Max Temperature', color='gold', kde=True)
plt.xlabel('Maximum Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Maximum Temperature')
plt.show()
```

```python
# This visualizes the distribution of Minimum Temperature through Histogram
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Min Temperature', color='plum', kde=True)
plt.xlabel('Minimum Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Minimum Temperature')
plt.show()
```

```python
# This visualizes the distribution of Corn through Histogram
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Corn', color='lightcoral', kde=True)
plt.xlabel('Annual Corn Yield')
plt.ylabel('Frequency')
plt.title('Distribution of Corn Yield')
plt.show()
```

```python
# This visualizes the distribution of Soybean through Histogram
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='Soybean', color='Skyblue', kde=True)
plt.xlabel('Annual Soybean Yield')
plt.ylabel('Frequency')
plt.title('Distribution of Soybean Yield')
plt.show()
```

```python
# This visualizes the correlation matrix of the variables in the dataset
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# This creates a correlation heatmap between the variables 
plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Climate and Crops Annual Data')

# This display the correlation heatmap
plt.show()
```

```python
# This creates a boxplot for Precipitation
variable_to_plot = 'Precipitation'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Precipitation', color='mediumseagreen')
plt.title(f'Boxplot of Precipitation')
plt.ylabel('Values')
plt.show()
```

```python
# This creates a boxplot for Minimum Temperature
variable_to_plot = 'Min Temperature'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Min Temperature', color='plum')
plt.title(f'Boxplot of Minimum Temperature')
plt.ylabel('Values')
plt.show()
```

```python
# This creates a boxplot for Maximum Temperature
variable_to_plot = 'Max Temperature'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Max Temperature', color='gold')
plt.title(f'Boxplot of Maximum Temperature')
plt.ylabel('Values')
plt.show()
```

```python
# This creates a boxplot for Corn
variable_to_plot = 'Corn'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Corn', color='lightcoral')
plt.title(f'Boxplot of Corn Yield')
plt.ylabel('Values')
plt.show()
```

```python
# This creates a boxplot for Soybean
variable_to_plot = 'Soybean'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Soybean', color='Skyblue')
plt.title(f'Boxplot of Soybean Yield')
plt.ylabel('Values')
plt.show()
```

```python
# This selects the variables for the boxplot 
selected_columns = ['Precipitation', 'Max Temperature', 'Min Temperature', 'Corn','Soybean']
data_subset = df[selected_columns]
# This creates the boxplot for the selected variables
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_subset, orient="v", palette="Set3", fliersize=8)
plt.title('Boxplots of Selected Variables')
plt.xlabel('Values')
plt.show()
```

```python
# This imports the library
import statsmodels.api as sm
# This provides the regression analysis of Corn with Climatic Variables
df['const'] = 1
X = df[['const', 'Precipitation', 'Max Temperature', 'Min Temperature']]
y = df['Corn']
model_corn = sm.OLS(y, X).fit()
print("Regression Analysis for Corn:")
print(model_corn.summary())
```

```python
# This provides the regression analysis of Soybean with Climatic Variables
df['const'] = 1
X = df[['const', 'Precipitation', 'Max Temperature', 'Min Temperature']]
y = df['Soybean']
model_soybean = sm.OLS(y, X).fit()
print("Regression Analysis for Soybean:")
print(model_soybean.summary())
```

```python
# This provides Scatter plot for Precipitation vs. Corn yield
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Precipitation', y='Corn', color='blue', label='Precipitation')
sns.regplot(data=df, x='Precipitation', y='Corn', color='blue', scatter=False)
plt.title('Scatter Plot of Precipitation vs. Corn Yield')
plt.xlabel('Precipitation')
plt.ylabel('Corn Yield')

# This provides Scatter plot for Max Temperature vs. Corn yield
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Max Temperature', y='Corn', color='green', label='Max Temperature')
sns.regplot(data=df, x='Max Temperature', y='Corn', color='green', scatter=False)
plt.title('Scatter Plot of Max Temperature vs. Corn Yield')
plt.xlabel('Max Temperature')
plt.ylabel('Corn Yield')

# This provides Scatter plot for Min Temperature vs. Corn yield
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Min Temperature', y='Corn', color='red', label='Min Temperature')
sns.regplot(data=df, x='Min Temperature', y='Corn', color='red', scatter=False)
plt.title('Scatter Plot of Min Temperature vs. Corn Yield')
plt.xlabel('Min Temperature')
plt.ylabel('Corn Yield')

plt.show()
```
